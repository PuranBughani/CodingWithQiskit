## Getting started with QC ##


```python
from qiskit import *
from qiskit.visualization import plot_histogram


```

```python
n = 8
n_q = n
n_b = n
qc_output = QuantumCircuit(n_q,n_b)
```

```python
# nqubits = 2
# circ = QuantumCircuit(nqubits, nqubits)
# circ.h(0)
# circ.cx(0,1)
# circ.cx(1,2)
# circ.measure([0,1],[0,1])
```

```python
for j in range(n):
    qc_output.measure(j,j)
```

```python
# All 8 bits are defaulted by 0 so making it 0. measure has attached to all  
qc_output.draw(output='mpl')
```

```python
#Check the value is one for all bit is 0 making in it 0 

counts = execute(qc_output,Aer.get_backend('qasm_simulator')).result().get_counts()
plot_histogram(counts)
```
