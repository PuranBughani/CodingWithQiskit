## QuantumTeleportation *(QT)* Algorithm

```python
# Moving Qubit(Qb) information from 1 to another(Qb) - teleporation  
```

```python
# teleportation require following gates 

# QuantumTeleportation(QT) is moving QuantumInformation(QI) from one Qb to another
# Entanglement (En) is required for teletrnsportation  
# Superpostion(Sp) is not required 
```

```python
from qiskit import *
```

```python
circ = QuantumCircuit(3,3)
```

```python
%matplotlib inline
circ.draw(output='mpl')
```

## X gate - property of flipping the state to , and vice versa 

The Pauli X gate has the property of flipping the
state to , and vice versa. It is equivalent to Rx for the angle




```python
# apply x gate in q0 so that state change from 0 to 1 

circ.x(0)
circ.draw(output='mpl')
```

```python
# attach circit barrier circuit.barrier() from 0 to 2 quantum state
# and it will remain ????? 
circ.barrier()
circ.draw(output='mpl')
```

### H   gate - useful for making superpositions ##
The H or Hadamard gate rotates the states and to and , respectively.
It is useful for making superpositions. As a Clifford gate, it is useful 
for moving information between the x and z bases.



```python

circ.h(1)
circ.cx(1,2)
circ.draw(output='mpl')
```

```python
# teleportation protocol to be added with C-NOT from Qb 1,2 and Hadamard gate at Qb0


circ.cx(0,1)
circ.h(0)
```

```python
circ.draw(output='mpl')
```

```python
# Need to add messurement at Qb 0 and 1, 
circ.barrier()
circ.measure([0,1], [0,1])
circ.draw(output='mpl')
```

```python
# 2 more gates are required here to complete TT (0 to 2)
circ.barrier()
circ.cx(1,2)
circ.cz(0,2)
circ.draw(output='mpl')

```

```python

# adding measure to the say copy the value of Qb 2 to Cb 2
circ.measure(2,2)
# setup backend simulator
simulator =  Aer.get_backend('qasm_simulator')
#execute() circut + get result() in vaiable result 
result = execute(circ, backend=simulator, shots = 1024).result()
#store the counts of result in the count variable to create histogram
counts = result.get_counts() 

# import plot_histogram from qiskit tools visulalization
from qiskit.tools.visualization import plot_histogram
from qiskit.tools.visualization import plot_bloch_multivector

# finally plot the histogram chart of result count recevied after executeion of TT circuit 
plot_histogram(counts) 
```
